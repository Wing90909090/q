```python
import pennylane as qml
import numpy as np
import matplotlib.pyplot as plt

# Set the number of steps and the size of the grid
num_steps = 10
grid_size = 21

# Initialize the quantum device and register
dev = qml.device("default.qubit", wires=3, shots=1)

@qml.qnode(dev)
def circuit(theta, phi, step):
    # Prepare initial state
    qml.Hadamard(wires=0)
    qml.Hadamard(wires=1)

    # Apply controlled rotation gates based on coin state
    for i in range(step):
        qml.CRZ(theta, wires=[0, 1])
        qml.CRY(phi, wires=[0, 1])

        # Shift the position of the walker
        qml.CSWAP(wires=[2, 1, 0])

    # Measure the final positions
    return [qml.sample(qml.PauliZ(wires=i)) for i in range(2)]

# Initialize the grid positions
x = np.zeros(num_steps+1)
y = np.zeros(num_steps+1)

# Perform the quantum walk
for step in range(num_steps):
    theta = 2 * np.pi * np.random.random()
    phi = np.pi * np.random.random()

    result = circuit(theta, phi, step)
    x[step+1] = x[step] + result[0][0]
    y[step+1] = y[step] + result[1][0]

# Plot the quantum random walk trajectory
plt.figure(figsize=(6, 6))
plt.plot(x, y, linestyle='-', marker='o', color='blue', label='Trajectory')
plt.scatter(x[0], y[0], color='green', label='Starting Point')
plt.scatter(x[-1], y[-1], color='red', label='Ending Point')
plt.title("Quantum Random Walk in 2D")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.xlim([-grid_size, grid_size])
plt.ylim([-grid_size, grid_size])
plt.grid(True)
plt.legend()
plt.show()
```
